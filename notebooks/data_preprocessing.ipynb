```python
# Clean column names by removing spaces, replacing special characters, etc.
df.columns = df.columns.str.strip()  # Remove leading and trailing spaces
df.columns = df.columns.str.replace('%', 'percent', regex=False)  # Replace '%' with 'percent'
df.columns = df.columns.str.replace(' ', '_', regex=False)  # Replace spaces with underscores
df.columns = df.columns.str.replace('-', '_', regex=False)  # Replace hyphens with underscores
df.columns = df.columns.str.replace('(', '', regex=False).str.replace(')', '', regex=False)  # Remove parentheses

print(df.columns.tolist())
```

```python
columns_to_convert = [
    'Out_of_pocket_health_expenditure',
    'Population', 
    'Population:_Labor_force_participation_percent', 
    'Tax_revenue_percent', 
    'Total_tax_rate', 
    'Unemployment_rate', 
    'Urban_population',
    'Minimum_wage',
    'Forested_Area_percent',
    'Gasoline_Price',
    'GDP',
    'Gross_primary_education_enrollment_percent',
    'Gross_tertiary_education_enrollment_percent',
    'Co2_Emissions',
    'CPI',
    #'CPI_Change_percent',
    'Agricultural_Land_percent',
    'Land_AreaKm2',
    'Armed_Forces_size'
]

# Check for non-numeric values in each column before conversion
for col in columns_to_convert:
    non_numeric = df[col].apply(lambda x: isinstance(x, str) and not x.replace('.', '', 1).isdigit())
    if non_numeric.any():
        print(f"Non-numeric values in column {col}:")
        print(df.loc[non_numeric, col])

# Example: Cleaning up non-numeric characters (commas, spaces, etc.)
for col in columns_to_convert:
    df[col] = df[col].str.replace(',', '', regex=False)  # Remove commas
    df[col] = df[col].str.replace(' ', '', regex=False)  # Remove spaces
    df[col] = df[col].str.extract('(\d+\.?\d*)')  # Extract only numeric parts
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric

# Process CPI_Change_percent separately to retain negative values
df['CPI_Change_percent'] = df['CPI_Change_percent'].str.replace(',', '', regex=False)
df['CPI_Change_percent'] = df['CPI_Change_percent'].str.replace(' ', '', regex=False)
df['CPI_Change_percent'] = df['CPI_Change_percent'].str.extract(r'(-?\d+\.?\d*)')[0]  # Retain negative values
df['CPI_Change_percent'] = pd.to_numeric(df['CPI_Change_percent'], errors='coerce')
```

```python
# Function to convert alpha-2 code to continent
def get_continent_from_alpha2(alpha2_code):
    try:
        # Convert alpha-2 code to continent code
        continent_code = pc.country_alpha2_to_continent_code(alpha2_code)
        # Convert continent code to continent name
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except Exception as e:
        return None  # Return None if the country code is not found or valid

# Apply the function to create the continent column using the Abbreviation column
df['Continent'] = df['Abbreviation'].apply(get_continent_from_alpha2)

# Verify the results
print(df[['Abbreviation', 'Continent']].head(10))


        
```
